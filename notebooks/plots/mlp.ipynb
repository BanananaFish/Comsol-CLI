```python
import matplotlib.pyplot as plt

# 
layers = [5, 8, 8, 3]  # 
layer_x = [0, 1, 2, 3]  #  x 

# 
fig, ax = plt.subplots(figsize=(8, 6))
ax.axis('off')  # 

# 
for i, num_nodes in enumerate(layers):
    for j in range(num_nodes):
        # 
        x, y = layer_x[i], j - num_nodes / 2
        ax.add_patch(plt.Circle((x, y), radius=0.2, color='lightgray', ec='black', zorder=2))
        
        # 
        if i > 0:
            for k in range(layers[i-1]):
                x_prev, y_prev = layer_x[i-1], k - layers[i-1] / 2
                ax.plot([x_prev, x], [y_prev, y], color='gray', zorder=1)

# 
ax.text(-0.5, 0, 'Input\nLayer', ha='center', fontsize=12)
ax.text(1, 4, 'Hidden Layers', ha='center', fontsize=12)
ax.text(3.5, 0, 'Output\nLayer', ha='center', fontsize=12)

plt.show()
```
