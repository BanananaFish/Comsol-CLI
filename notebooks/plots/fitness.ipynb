```python
from comsol.ga import fit
from comsol.utils import Config
import os
from config import fontsize, ticksize
os.chdir("E:/code/comsol")
exps = "exports/slit-9-1/params_del_bad.csv"
config = "config/cell_local.yaml"
ckpt = r"ckpt\2024.11.12_19.33.43\earlystop_best_0.000514.pth"

cfg = Config(config)
fitnesses = fit(ckpt, exps, cfg)
```

```python
from matplotlib import pyplot as plt
colors = {
    "blue": "#2E54A1",
    "green": "#249087",
    "red": "#FF0000"
}
plt.rcParams["legend.fontsize"] = 'large'
plt.rcParams["font.family"] = 'Times New Roman'
plt.plot(fitnesses, color=colors["blue"])
plt.tick_params(axis='both', labelsize=ticksize)
# plt.title("Fitnesses", fontsize=fontsize)
plt.xlabel("Generations", fontsize=fontsize)
plt.ylabel("Fitness value", fontsize=fontsize)
plt.tight_layout()
plt.savefig("notebooks/plots/figures/fig3d_fitness.pdf")
plt.show()
```
